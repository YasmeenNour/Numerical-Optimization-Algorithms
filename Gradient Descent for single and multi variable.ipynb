## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate.

```python
theta_0=0
theta_1=0
alpha=0.0001
h_x=0
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
h_x=theta_0+theta_1*X
print(f"h(x): {h_x}")
```

### Step3: Calculate the MSE Cost function 𝑱(theta_0,theta_1 ).
<b>Note:</b> You better use either dot product or norm square of the error vector.

```python

error_vector=np.array(13)
error_vector_squared=np.array(13)

#y^-y
error_vector=h_x-y
print("Error Vector: ")
print(error_vector)
print(" ")
error_vector_squared=pow(error_vector,2)
j=1/(2*m)*np.sum(error_vector_squared)
print(f"j = {j}")
```

### Step4: Calculate the gradient components for theta_0 and theta_1.
<b>Note:</b> You can use the error vector calculated in the previous step.

```python
d_theta_0=(1/m)*np.sum(error_vector)
d_theta_1=(1/m)*np.dot(error_vector,X)

print(f"d_theta_0 : {d_theta_0}")
print(f"d_theta_1 : {d_theta_1}")


#gradient_norm
gradient_norm=np.sqrt(pow(d_theta_0,2)+pow(d_theta_1,2))

  
```

### Step5: Update the parameters (simultaneously).

```python
theta_0_new=theta_0-(alpha*d_theta_0)

theta_1_new=theta_1-(alpha*d_theta_1)

print(f"theta_0_new : {theta_0_new}")
print(f"theta_1_new : {theta_1_new}")
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.
#### The objective from this step is to combine all the previous steps and iterate untill you either achieve the maximum number of iterations or reach the minimum.

```python
theta_0=0
theta_1=0
alpha=0.01
h_x=0
```

```python
for i in range(0,1000):
  #calculate Y predict
  h_x=theta_0+theta_1*X
  error_vector=np.array(13)
  error_vector_squared=np.array(13)


  #calculate cost between y and y predict
  error_vector=h_x-y
  error_vector_squared=pow(error_vector,2)
  #calculate cost function
  j=1/(2*m)*np.sum(error_vector_squared)
  

  #calculate gradient components
  d_theta_0=(1/m)*np.sum(error_vector)
  d_theta_1=((1/m)*np.sum(error_vector*X))

  gradient_vector=np.array([[d_theta_0],[d_theta_1]])


  #gradient_norm
  gradient_norm=np.linalg.norm(gradient_vector)

  #update theta
  theta_0=theta_0-(alpha*d_theta_0)

  theta_1=theta_1-(alpha*d_theta_1)
 

  print(f"****************** Iteration {i} ********************\n")
  print(f"h(x): {h_x}")
  print(" ")

  print("Error Vector: ")
  print(error_vector)
  print(" ")

  print(f"j = {j}")
  print(" ")

  print(f"Gradient Vector:")
  print(f"{gradient_vector}\n")

  print("Gradient Vector Norm:")
  print(f"{gradient_norm}\n")

  print(f"theta_0_new : {theta_0}")
  print(f"theta_1_new : {theta_1}\n")

  

  
```

```python

for i in range(0,500):
  #calculate Y predict
  h_x=theta_0+theta_1*X
  error_vector=np.array(13)
  error_vector_squared=np.array(13)


  #calculate cost between y and y predict
  error_vector=h_x-y
  error_vector_squared=pow(error_vector,2)
  #calculate cost function
  j=1/(2*m)*np.sum(error_vector_squared)
  

  #calculate gradient components
  d_theta_0=(1/m)*np.sum(error_vector)
  d_theta_1=((1/m)*np.sum(error_vector*X))

  gradient_vector=np.array([[d_theta_0],[d_theta_1]])


  #gradient_norm
  gradient_norm=np.linalg.norm(gradient_vector) 

  print(f"****************** Iteration {i} ********************\n")
  print(f"h(x): {h_x}")
  print(" ")

  print("Error Vector: ")
  print(error_vector)
  print(" ")

  print(f"j = {j}")
  print(" ")

  print(f"Gradient Vector:")
  print(f"{gradient_vector}\n")

  print("Gradient Vector Norm:")
  print(f"{gradient_norm}\n")

  thetas_opt=[]
  if gradient_norm<=0.09:
    thetas_opt.append([theta_0,theta_1])

    print("****************** Training Report ********************\n")
    print(f"Gradient Descent converged after {i} iterations")
    print(f"theta_0_Opt :{thetas_opt[0][0]}")
    print(f"theta_1_Opt :{thetas_opt[0][1]}")
    print("Error Vector: ")
    print(error_vector)
    print(" ")
    print(f"Cost = {j}")

    print("h(x) = y_predict:")
    print(f"{h_x}\n")
    print("y_actual:")
    print(y)
    break


  #update theta
  theta_0=theta_0-(alpha*d_theta_0)
  theta_1=theta_1-(alpha*d_theta_1)

  print(f"theta_0_new : {theta_0}")
  print(f"theta_1_new : {theta_1}\n")

  

  
```

### Generalize your implementation to work for MultiVariables
#### Vectorize your implementation.
<b>Notes:</b> 
- You need to add column that contains ones to the data. This column represent x feature for theta_0.
- Use learninig rate = 0.0001.

```python
x0=np.ones([Xs.shape[0],1])
arr_x = np.concatenate([x0, Xs], axis=1)
m=data2.shape[0]
thetas=np.zeros([arr_x.shape[1] ,1])
alpha= 0.0001
h_x=0
thetas
```

```python
iter=0
epochs=[]

for i in range (0,5000):
  #calculate Y predict
  h_x= arr_x @thetas
  y2=y2.reshape(25,1)



  # calculate cost between y and y predict
  error_vector=h_x-y2


  # #calculate cost function
  #if the old data m=13 the j = 25780.73076923077
  #if the New m =25 the j is different
  j=1/(2*m )*np.sum(pow(error_vector,2))


  ##calculate gradient components
  ##check this step again
  gradient_components=(1/m)*np.dot(arr_x.T,error_vector)

  #calculate the gradient Norm
  gradient_norm=np.linalg.norm(gradient_components)



  print(f"****************** Iteration {i} ********************\n")
  print(f"h(x):\n {h_x}")
  print(" ")

  print("Error Vector: ")
  print(error_vector)
  print(" ")

  print(f"j = {j}")
  print(" ")

  print(f"Gradient Vector:")
  print(f"{gradient_components}\n")

  print("Gradient Vector Norm:")
  print(f"{gradient_norm}\n")

  thetas_opt=[]
  if gradient_norm<0.1:
    thetas_opt.append(gradient_components)

    print("****************** Training Report ********************\n")
    print(f"Gradient Descent converged after {i} iterations")
    print(f"thetas_Opt \n:{gradient_components}")
    print("Error Vector: ")
    print(error_vector)
    print(" ")
    print(f"Cost = {j}")

    print("h(x) = y_predict:")
    print(f"{h_x}\n")
    print("y_actual:")
    print(y2)
    iter=i
    break


  #update theta
 #calculate the new thetas
  thetas=thetas-alpha*gradient_components

  print(f"thetas_new :\n {thetas}")
  epochs.append(j)

  

  
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x1 + theta_2 * x2 + theta_3 * x3

```python
print(f"h(x) = y_predict:\n{h_x}")
print(f"y:\n{y2}")
```

### Plot loss vs. iterations

```python
plt.rcParams['figure.figsize'] = [10.8, 7.6]
plt.plot(range(0,iter),epochs,"-*")
plt.ylabel("loss")
plt.xlabel("Epochs")
plt.grid()
```
