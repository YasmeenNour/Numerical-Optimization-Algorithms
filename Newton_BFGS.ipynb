##### Notes:

```python
f = lambda x: x**2 - 15*x +1
Df = lambda x: 2*x -15

xp=np.arange(-44,60)
DDf = lambda x: 2*np.ones_like(x)
```

```python
def draw_plots(Xp,f,df,ddf):
  yp=f(Xp)
  df_y=df(Xp)
  ddf_y=ddf(Xp)

#fx
  plt.plot(Xp,yp)
  plt.title("f(x)=x^2-15x+1")
  plt.grid()
  plt.show()

  #df
  plt.plot(Xp,df_y)
  plt.title("Df(x)=2^x-15")
  plt.grid()
  plt.show()

  #ddf
  plt.plot(Xp,ddf_y)
  plt.title("DDf(x)=2")
  plt.grid()
  plt.show()
```

```python
draw_plots(xp,f,Df,DDf)
```

```python
# Gradient descent with learning rate = 0.01, x0=40, epsilon = 0.1
def gradient_descent_iterations(x ,alpha ,epsilon ,df=Df):
    print(f"GD Results with x0 ={x}, lr={alpha}, epsilon={epsilon}")
    iter =0
    gd = 1
    while gd>epsilon:
        gd = df(x)
        x = x - (alpha * gd)
        x_min = x

        iter += 1

    print(f"Found solution after {iter} iterations.\n")
    print(f"x_min = {x_min}")
    print(f"Gradient = {gd}")
    return gd,x_min  
```

```python
g1,x1=gradient_descent_iterations(40,0.01,0.1)
```

```python
# Newton's method without learning rate, x0=40, epsilon = 0.1
def newton_iteration(x  ,epsilon ,alpha=1,df=Df,ddf=DDf):
    print(f"Newton Results with x0 ={x}, epsilon={epsilon}")
    iter =0
    gd = df(x)/ddf(x)
    # gd_norm=np.linalg.norm(gd)
    while gd> epsilon:
        x = x -  alpha*gd
        x_min = x

        iter += 1
        gd=df(x)/ddf(x)
        # gd_norm=np.linalg.norm(gd)

    print(f"Found solution after {iter} iterations.\n")
    print(f"x_min = {x_min}")
    print(f"Gradient = {gd}")
    return gd,x_min
```

```python

g2,x_min=newton_iteration(40, 0.1)
```

```python
# Newton's method with learning rate = 0.8, x0=40, epsilon = 0.1
g3,x_min=newton_iteration(40,0.1,0.8)
```

```python
f_1 = lambda x: x**2  + y**2
Df_1 = lambda X: np.array([2*X[0],2*X[1]])
DDf_1 = lambda X: np.array([[2,0],
                         [0,2]])
```

```python
f_2 = lambda x: x**2 + x*y + y**2
Df_2 = lambda X: np.array([2*X[0]+X[1],2*X[1]+X[0]])
DDf_2 = lambda X: np.array([[2,1],
                        [1,2]])
```

```python
# First function
# Multi-variable Gradient descent with learning rate = 0.1,
# x0=(10,10), epsilon = 0.001
def multi_var_gd(x,alpha,epsilon,df=Df_1):
  iter =0
  gd=df(x)
  x = x - alpha*gd
  gd_norm=np.linalg.norm(gd)
  while gd_norm > epsilon:
      gd = df(x)
      xmin=x
      x = x - alpha * gd
      gd_norm=np.linalg.norm(gd)
      
      iter += 1
 
  print(f"Found solution after {iter} iterations.\n")
  print(f"xy_min = {xmin}")
  print(f"Gradient = {gd}")
  return gd,xmin

```

```python
x=[10,10]
multi_var_gd(x,0.1,0.001)


```

```python
# First function
# Multi-variable Newton's method without learning rate, 
# x0=(10,10), epsilon = 0.001
def multi_var_newton(x  ,epsilon ,alpha=1,df=Df_1,ddf=DDf_1):
  iter =0
  gd=df(x)
  # hessian= np.linalg.inv(ddf(x))
  # x = x - alpha * (hessian @ gd)
  gd_norm=np.linalg.norm(gd)

  while gd_norm>epsilon:
      gd = df(x)
      hessian= np.linalg.inv(ddf(x))
      x = x - alpha * np.dot(hessian,gd).T
      x_min = x
      iter += 1
      gd = df(x)
      gd_norm=np.linalg.norm(gd)


  print(f"Found solution after {iter} iterations.\n")
  print(f"xy_min = {x_min}")
  print(f"Gradient = {gd}")
  return gd,x_min
```

## BFGS
- Implement the <b>BFGS</b> as a function and find the minimum of the following function:
    - $ùíá(ùíô)=ùíô^ùüê+y^2$
    - $ùíá(ùíô)=ùíô^ùüê+xy+y^2$  
    - $ùíá(ùíô)=ùíô^ùüêy$ 
- The stop condition is grad < epsilon.
- Try different epsilons and initial conditions.
- Compare the results from Gradient descent and Newton's method.

```python
def multi_var_BFGS(x0, xk_1, epsilon, alpha=1,df=Df_1):

  # initialize b_1
  b_1 = np.eye(len(x0))

  # calculatefirst dervative
  x0=np.asarray(x0).reshape(2,-1)
  xk_1=np.array(xk_1).reshape(2,-1)
  gd = df(x0)
  gd_1=df(xk_1)

  gd_norm = np.linalg.norm(gd)

  d_x = np.asarray(x0) - np.asarray(xk_1).reshape(2, -1)
  y_t = df(x0) - df(xk_1).reshape(2, -1)




  # y_t = np.asarray(y_t)

  iter = 0
  i=np.eye(len(x0))

  print("*******")
  print(f"xn_1:\n {xk_1}")
  print(f"xn:\n {x0}")
  print(f"dfxn_1:\n {gd_1}")
  print(f"dfxn:\n {gd}")
  print("norm", gd_norm)
  print("*************")

  while gd_norm > epsilon:

    gd = df(x0).reshape(2,-1)
    gd_1=df(xk_1).reshape(2,-1)
    b_1 = ( i - ( (d_x @ y_t.T)/(y_t.T@d_x) ) )@ b_1 @( i - ( (y_t @d_x.T)/(y_t.T@d_x) ))+((d_x @ d_x.T)/(y_t.T@d_x))
    print(f"B inverse : \n{b_1}")

    xk_1 = x0
    x0 = x0 - alpha*(np.dot(b_1 , gd))

    x_min = x0
    d_x = x0 - xk_1
    y_t = df(x0) - df(xk_1)
    y_t = np.asarray(y_t)
    iter += 1
    gd = df(x0).reshape(2,-1)
    gd_1 = df(xk_1).reshape(2, -1)
    gd_norm = np.linalg.norm(gd)

    print("*************\n")
    print(f"xn_1:\n {xk_1}\n")
    print(f"xn:\n {x0}\n")
    print(f"dfxn_1:\n {gd_1}\n")
    print(f"dfxn:\n {gd}\n")
    print("norm", gd_norm)
    print("*************\n")

  print(f"Found solution after {iter} iterations.\n")
  print(f"xy_min = {x_min}\n")
  print(f"Gradient = {gd}\n")
  return gd,x_min

```
